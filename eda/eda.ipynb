```python
def hist_plot_user_engagement(df):
    x_ticks = np.arange(0,325,25)
    x = df['appopen']
    y = df['1w_open'].dropna()
    z = df['1d_open'].dropna()
    bins = 400
    plt.hist(x, bins, alpha=0.5, label='Total')
    plt.hist(y, bins, alpha=0.5, label='First Week')
    plt.hist(z, bins, alpha=.5, label='First 2 Days')

    plt.legend(loc='upper right')
    plt.xlabel("Num of Times App Opened")
    plt.ylabel("Num of Users")
    plt.title("User App Engagement")

    plt.ylim(0,1000)

    plt.xlim(0 , 325)

    plt.xticks(x_ticks)
```

```python
hist_plot_user_engagement(df)
```

```python
def plot_10_users_ivan(df,bg_values, df_s):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    user_ids = df_s.user_id.values

    for user_id in user_ids:
        sub_df = bg_values[bg_values.user_id == user_id]

        sub_df.sort_values(by = ['days_f0bg', 'bg'], axis=0, ascending=[True, True], inplace=True)
        sub_df = sub_df[sub_df.bg_appopen>0]
        sub_df= sub_df.dropna()
        sub_df['bg']= sub_df['bg']*1.0/sub_df['bg'][0:1].values

        x = sub_df.days_f0bg
        y = sub_df.bg
    
        #plt.scatter(x = x, y = y, c = np.random.rand(3,1))

        sub_df.plot(x = 'days_f0bg', y = 'bg', c = np.random.rand(3,1), ax = ax, kind = 'line', label = 'user {}'.format(user_id))

    plt.legend(loc='upper right', prop={'size':5})
    plt.xlabel("app age")
    plt.ylabel("BG Levels ")
    plt.title("BG Changes vs Time for Low Risk Users")

    plt.xlim(0,120)
    #plt.ylim(0,400)
    plt.show()

```
